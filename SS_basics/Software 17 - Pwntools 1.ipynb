### Software 17 - Pwntools 1

```python
#!/usr/bin/env python3

from pwn import *

def main():
    HOST = "software-17.challs.olicyber.it"
    PORT = 13000
    r = remote(HOST, PORT)

    r.sendline(b"a")  # Avvia la challenge

    for step in range(10):
        r.recvuntil(b"numeri\n")
        numbers_line = r.recvline().decode().strip()
        print(f"[DEBUG] Numeri: {numbers_line}")

        try:
            numbers = eval(numbers_line)
            result = sum(numbers)
            print(f"[DEBUG] Somma: {result}")
            r.sendline(str(result).encode())
        except Exception as e:
            print(f"[ERROR] {e}")
            break

    # Mostra il risultato finale (flag)
    final_output = r.recvall().decode()
    print(final_output)

if __name__ == "__main__":
    main()
```
