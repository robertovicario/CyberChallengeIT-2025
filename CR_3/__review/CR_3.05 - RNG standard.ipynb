### CR_3.05 - RNG standard

```python
from pwn import *
import randcrack

# Connessione remota
HOST = "srng.challs.cyberchallenge.it"
PORT = 9064
r = remote(HOST, PORT)

rc = randcrack.RandCrack()

# Fase 1: Ottieni 624 numeri
for i in range(624):
    r.recvuntil(b"What do you want to do?")
    r.sendline(b"2")  # Give up
    r.recvuntil(b"My number was ")
    number_line = r.recvline()
    num = int(number_line.strip())
    print(f"[{i+1}/624] Numero: {num}")
    rc.submit(num)  # Inserisci numero in randcrack

# Fase 2: Predict il 625° numero
predicted = rc.predict_getrandbits(32)
print(f"[+] Numero predetto: {predicted}")

# Fase 3: Indovina il 625° numero
r.recvuntil(b"What do you want to do?")
r.sendline(b"1")  # Guess
r.recvuntil(b"> ")
r.sendline(str(predicted).encode())

# Ricevi la flag
response = r.recvline()
if b"CCIT{" in response:
    print(f"[+] FLAG: {response.decode().strip()}")
else:
    print(f"[-] Nessuna flag: {response.decode().strip()}")

r.close()
```
