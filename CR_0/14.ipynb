````python
from Crypto.Hash import SHA3_384

msg = 'hash_me_pls'

sha = SHA3_384.new()
sha.update(msg.encode('utf-8'))
sha.hexdigest()
````

````python
from Crypto.Hash import HMAC, SHA224

msg = 'La mia integrità è importante!'
key = '8d405ea5c668f53cb2eecd20dbe2cb28cd6b2e9b3bdc585a086bc2ed3e79a0ff'

hmac = HMAC.new(bytes.fromhex(key), digestmod=SHA224)
hmac.update(msg.encode('utf-8'))
hmac.hexdigest()
````
