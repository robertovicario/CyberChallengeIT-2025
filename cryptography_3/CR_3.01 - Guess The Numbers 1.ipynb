### CR_3.01 - Guess The Numbers 1

````python
m = 1076867677
c = 1265354953
n = 2147483647
s = 1862611659

def generate_sequence(m, c, n, s, count=50):
    sequence = []
    current_value = s
    
    for _ in range(count):
        next_value = (m * current_value + c) % n
        sequence.append(next_value)
        current_value = next_value
    
    return sequence

generate_sequence(m, c, n, s, count=50)
````
