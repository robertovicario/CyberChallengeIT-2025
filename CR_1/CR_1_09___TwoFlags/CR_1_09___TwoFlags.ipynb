### CR_1.09 - TwoFlags

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

img1 = Image.open('./flag_enc.png')
img2 = Image.open('./notflag_enc.png')

npImg1 = np.array(img1) * 255
npImg2 = np.array(img2) * 255
result = np.bitwise_xor(npImg1, npImg2)

plt.imshow(result)
plt.axis('off')
plt.show()
```
